Random Forest Regression

```python
import pandas as pd
import numpy as np

data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/MachineLearning/Day5/Position_Salaries.csv')
data
```

```python
X = data.iloc[:,1].values.reshape(-1,1)
Y = data.iloc[:,2].values
```

```python
# Fitting Random Forest Regression to the dataset
# import the regressor
from sklearn.ensemble import RandomForestRegressor

# create regressor object
regressor = RandomForestRegressor(n_estimators = 100, random_state = 0)

# fit the regressor with x and y data
regressor.fit(X, Y)
```

```python
Y_pred = regressor.predict(np.array([6.6]).reshape(-1, 1)) # test the output by changing values

Y_pred
```

```python
import matplotlib.pyplot as plt

X_grid = np.arange(min(X), max(X), 0.01)
X_grid = X_grid.reshape((len(X_grid), 1))
plt.scatter(X, Y, color = 'red')
plt.scatter([6.6], Y_pred, color = 'green')
plt.plot(X_grid, regressor.predict(X_grid), color = 'blue')
plt.title('Truth or Bluff (Random Forest Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```
